```
## NEWS Rack
import re
import requests
from bs4 import BeautifulSoup

urls = []
def scrape(URL):
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    b = soup.findAll(attrs={"class": "newsnavbar"})
    txt = b[1].text
    c = txt.split(' ')
    rst = c[4].split(',')
    total_results = ('').join(rst)
    # print(total_results)
    total_results = int(float(total_results))
    # print(soup.title)
    for link in soup.find_all('a', href=re.compile(r"^https://")):
        # print(link.get('href'))
        urls.append(link.get('href'))
        # print(urls)
    print(len(urls))
    return total_results

URL = 'http://newsrack.in/archive/stories/news_bipp/idp_ag/9?start=1'
tr = scrape(URL)
# print(URL)
v = 21
pg = tr/20
# print(pg)
pg = int(pg)+1
while pg > 0:
    # print(pg,'pg')
    URL = "http://newsrack.in/archive/stories/news_bipp/idp_ag/9?start="+str(v)
    # print(URL)
    # scrape(URL)
    pg = pg -1
    v += 20
print(urls)
```
